```python
target = 'species'
encode = ['sex', 'island']
```

```python
for col in encode:
    dummy = pd.get_dummies(df[col], prefix = col) 
    # we applied pd.get_dummies() to the feature columns namely 'species' & 'Island' as 
    # thery are categorical features so we convert them into dummy variables (binary 0/1)
    df = pd.concat([df,dummy],axis=1)
    del df[col]
```
